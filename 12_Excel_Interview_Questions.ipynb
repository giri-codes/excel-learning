### *(Final Interview Revision)*


## **1. What is Microsoft Excel?**
Microsoft Excel is a spreadsheet application used to store, organize, analyze, and visualize data using rows, columns, formulas, and charts.


## **2. What is a cell in Excel?**
A cell is the intersection of a row and a column where data is entered, such as A1 or B5.


## **3. What is a workbook?**
A workbook is an Excel file that contains one or more worksheets.


## **4. What is a worksheet?**
A worksheet is a single sheet within a workbook used to enter and analyze data.


## **5. What is a formula in Excel?**
A formula is an expression that performs calculations in Excel and always starts with `=`.


## **6. What is a function in Excel?**
A function is a predefined formula that performs a specific calculation, such as SUM or AVERAGE.


## **7. Difference between formula and function?**
A formula is user-defined, while a function is predefined and has a fixed syntax.


## **8. What is SUMIF and why is it used?**
SUMIF adds values based on a given condition and is used for conditional summation.


## **9. What is COUNTIF?**
COUNTIF counts the number of cells that meet a specific condition.


## **10. What is VLOOKUP?**
VLOOKUP searches for a value in the first column of a table and returns a corresponding value from another column.


## **11. What are the limitations of VLOOKUP?**
VLOOKUP works only left to right and requires a column index number.


## **12. What is XLOOKUP?**
XLOOKUP is an advanced lookup function that replaces VLOOKUP and works in both directions.


## **13. Difference between VLOOKUP and XLOOKUP?**
VLOOKUP requires a column index and works only left to right, whereas XLOOKUP is more flexible and accurate.


## **14. What is Conditional Formatting?**
Conditional Formatting changes the appearance of cells based on rules to highlight important values.


## **15. What is a Pivot Table?**
A Pivot Table is used to summarize large amounts of data quickly without writing formulas.


## **16. What is a Dashboard in Excel?**
A dashboard is a single-page report that shows KPIs and charts for quick decision-making.


## **17. What are KPIs?**
KPIs (Key Performance Indicators) are important metrics used to measure performance.


## **18. Difference between chart and dashboard?**
A chart shows one insight, while a dashboard combines multiple charts and KPIs in one view.
